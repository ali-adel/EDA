```python
import numpy as np
import scipy as sp # have theotical statistics
from scipy.stats import randint , norm # uniform distribution / normal distribution
import matplotlib.pyplot as plt
```

# Distribution for continues random variables

probabilty denisty function (pdf)

image.png

```python
x = np.linspace(-5,5) # generate numbers from -5 t 5
normal_dist = norm.pdf(x) # it just give a number to can make the curve of the pdf by formula above
normal_dist
```

```python
plt.plot(x,normal_dist) # this just represent the pdf of normal distribution
```

comulative distribution function (cdf)


```python
cdf_values = norm.cdf(x) # it represent the probability of all possible values in pdf (by integrate area under the curve)
cdf_values
```

```python
plt.plot(x,cdf_values) # it increase untill reach value 1 as the whole area under the pdf curve is equal to one
```
