```python
import numpy as np
import seaborn as sns
from scipy.stats import bernoulli
```

# Bernoulli Distribution

image.png

**any distribution that have a binary representaion or two possible values**

 **p -> probability of success**

**1 - p or (q) -> probability of failure**

# let's see the theoritical distribution how it is look like

```python
ber_succ = bernoulli.pmf(1,0.8)
ber_succ
# so it is exactly 80% for success
```

```python
ber_fail = bernoulli.pmf(0,0.8)
ber_fail
# so it is exactly 20% for failure
```

```python
bernoulli.stats(0.8,moments='mv')
```

**so here the mean is probability of success (m = p)**

```python
var = ber_fail *  ber_succ
var
```

**variance here = p (probability of success) * q (probability of failure)**

```python
x=[0,1]
y = [ber_fail,ber_succ]
sns.barplot(x=x,y=y)
y
```

# so let's take a random sample from Bernoulli Distribution

```python
rand_brenolli = np.random.binomial(n=1,p=0.8,size=10)
cnt = sns.countplot(x=rand_brenolli)

print(len(rand_brenolli[rand_brenolli==1]),len(rand_brenolli[rand_brenolli==0]))
# it is almost 90% for success and 10% for failure so like we see it is far from the  Bernoulli Distribution that is 80% for success
```

```python
rand_brenolli = np.random.binomial(n=1,p=0.8,size=100000)
cnt = sns.countplot(x=rand_brenolli)

print(len(rand_brenolli[rand_brenolli==1]),len(rand_brenolli[rand_brenolli==0]))
# it is almost 80% for success and 20% for failure
```

# so what if we increase the sample size

```python
rand_brenolli = np.random.binomial(n=1,p=0.8,size=10000000)
cnt = sns.countplot(x=rand_brenolli)

print(len(rand_brenolli[rand_brenolli==1]),len(rand_brenolli[rand_brenolli==0]))
#it is depending on sample size and the sample taken so it looks a near for theoritical one
```

# so as we notice from the examples when you increases the sample size you actually be so near from your population distribution shape and this is for all distributions not only Bernoulli Distribution
