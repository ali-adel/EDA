# Theoritical

# 1. p(s) = 0.8

```python
n = 5 # number of exams
p = 0.8 # prob of success
x = np.arange(0,6) #RV(x) number of passed exams

prob = binom.pmf(x,n,p) # calculate prob

first_dist = sns.barplot(x=x,y=prob)
#as we see here it is left sekewed
```

# 2. p(s) = 0.5

```python
n = 5 # number of exams
p = 0.5 # prob of success
x = np.arange(0,6) #RV(x) number of passed exams

prob = binom.pmf(x,n,p) # calculate prob

second_dist = sns.barplot(x=x,y=prob)
#as we see here it is normal distributions
```

# 3. p(s) = 0.2

```python
n = 5 # number of exams
p = 0.2 # prob of success
x = np.arange(0,6) #RV(x) number of passed exams

prob = binom.pmf(x,n,p) # calculate prob

third_dist = sns.barplot(x=x,y=prob)
#as we see here it is right sekewed
```

# note : regrdeless the p(s) if n => number of exams (num of occurences(experiment))is large the distribution is almost normal (both have almost the same shape )
#this happens because of central limit theorem (will discussed later)

```python
data = np.random.binomial(n=500 , p=0.8 , size = 500  )
c1 = sns.countplot(x=data)
#as we see here it is normal and p = 0.8
```
