```python
import numpy as np
from scipy.stats import uniform
from scipy.stats import expon
import seaborn as sns
import matplotlib.pyplot as plt
```

#Uniform Continuous Distribution


image.png

# Theoritical

```python
a = 10
b = 30
mu,sigma = uniform.stats(loc=a,scale=b-a,moments='mv')
print("mean : ",mu ,"sigma : " ,sigma)
```

# p(15 < x < 20)

```python
p1 = uniform.cdf(20,loc=a,scale=b-a) - uniform.cdf(15,loc=a,scale=b-a)
print(p1)
```

# p(10 < x < 30)

```python
p2 = uniform.cdf(30,loc=a,scale=b-a) - uniform.cdf(10,loc=a,scale=b-a)
print(p2)
```

# PDF and CDF

```python
a = 10
b = 30
x = np.arange(a,b+1)
pdf = uniform.pdf(x,loc=a,scale=b-a)
cdf = uniform.cdf(x,loc=a,scale=b-a)

plt.plot(x,pdf)
plt.show()
plt.plot(x,cdf)
```

# Practical

# random sample with size = 10

*    **looks like far away from the uniform distribution as the sample size is very small**

```python
random_dist = np.random.uniform(a,b,size=10)
print(random_dist.mean() , random_dist.var())
sns.histplot(random_dist)
```

# sample with size = 10000000

#Exponential Distribution

# Theoritical

# it is a special case of gamma distribution it is  used to measure the time untill specific event happen and used most of the time in measure the time of machine failure or customer arrival

image.png



*  Mean = 1 / lambda
*   std = 1 / lambda




# for exp if we have lambda = 2 customer/min in a restaurant and i want to know the probability of one customer arrival in this lambda in a specific range of time so when the lambda increase the mean time (mean) is decrease and the rate of the distribution increased so when lambda is high it has greater probability than other to have less time for customer arrival

# Theoritical

```python
lambda_1 = 1 /50000
lambda_2 = 1 /20000

m1 = expon.stats(loc=0 , scale=1/lambda_1,moments='m')
m2 = expon.stats(loc=0 , scale=1/lambda_2,moments='m')
print("mean_1 : ", m1 , "mean_2 : " , m2)
```

```python
x = np.linspace(1,200000)
p1 = expon.pdf(x,loc=0 , scale=1/lambda_1)
p2 = expon.pdf(x,loc=0 , scale=1/lambda_2)
plt.plot(x,p1)
plt.plot(x,p2)
```

# As we see the p2 (higher lambda ) have the higher rate

# Practical

# with size = 10

```python
lambda_1 = 1 /50000
lambda_2 = 1 /20000

p1 = np.random.exponential(scale=1/lambda_1,size=10)
p2 = np.random.exponential(scale=1/lambda_2,size=10)

print(p1.mean())
print(p2.mean())
sns.histplot([p1,p2],kde=True)
```

# with size = 10000

```python
lambda_1 = 1 /50000
lambda_2 = 1 /20000

p1 = np.random.exponential(scale=1/lambda_1,size=10000)
p2 = np.random.exponential(scale=1/lambda_2,size=10000)

print(p1.mean())
print(p2.mean())
sns.histplot([p1,p2],kde=True)
```
