# what if i have the comulative value (cdf) (integration from infinity till the value) and i want the point that have this comulative value ?

*  **We use ppf function as it the inverse of the cdf that exploit the symmetry of the normal distribution and gives you the value you want**



```python
norm.cdf(1) # THIS GIVES YOU THE VALUE FROM -INF TILL 1
```

```python
norm.ppf(0.8413447460685429) # this gives you the value that have the cdf you give
```

# what about the 95% percent you can exploit it

```python
norm.ppf(0.5  + (0.95/2))
```

```python
D = np.random.normal(loc=0 , scale = 1,size=1000) # by default m=0 and sigma(std)=1
print(D.mean())
print(D.std())
# so as the size still small it's almost near to theoritical mean and std
```
