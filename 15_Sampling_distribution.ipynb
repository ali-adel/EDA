# Info about the sampling distribution

```python
sample_means = [9.5,11.5,12,12,12.5,13,13.5,14,14.5,14.5,15.5,16,16.5,17,18]
df = pd.DataFrame(sample_means, columns=['Sample Means'])

# Calculate the mean of the sample means
mean_value = df['Sample Means'].mean()

# Create histplot
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Sample Means', discrete=True,kde=True)

# Add a vertical line at the mean value
plt.axvline(x=mean_value, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_value:.2f}')

# Add title and labels
plt.title('Histogram of Sample Means with Mean Line')
plt.xlabel('Sample Means')
plt.ylabel('Count')
plt.legend()

# Display the plot
plt.show()
```

#**Note** :  if we find that the probability of sample mean very small or the mean fall after 3 std (P(x' > 3)) so we have around 3% that this sample doesn't follow this distribution or can't represent the population as 97% of means should fall within 3std but this mean fall after that
