# **EX** : Apply CLT on Exponential distribution

# Recap : Exponential distribution shape

```python
lambda_= 3
D = np.random.exponential(1/lambda_,size=1000)
sns.histplot(D,kde=True)
```

# so let's take 10000 sample for example and watch what happen for the sampling distribution with different sample size to see the CLT effect

# with sample size or n = 1

 *  we notice that it follows the popultion distribution and the sampling distribution mean very close to population mean (regardeless the sample size n)




```python
means =[]
for i in range(10000):
  lambda_= 3
  D = np.random.exponential(1/lambda_,size=1)
  mu = D.mean()
  means.append(mu)
print("population mean : ",1/lambda_)
print("sampling distribution mean : ",np.mean(means))
sns.histplot(means,kde=True)
```

# with sample size or n = 30

  *  we notice that it begins to follow the normal distribution







```python
means =[]
for i in range(10000):
  lambda_= 3
  D = np.random.exponential(1/lambda_,size=30)
  mu = D.mean()
  means.append(mu)
print("population mean : ",1/lambda_)
print("sampling distribution mean : ",np.mean(means))
sns.histplot(means,kde=True)
```

# with sample size or n > 30
  * we notice that it follows the exact normal distribution





```python
means =[]
for i in range(10000):
  lambda_= 3
  D = np.random.exponential(1/lambda_,size=100000)
  mu = D.mean()
  means.append(mu)
print("population mean : ",1/lambda_)
print("sampling distribution mean : ",np.mean(means))
sns.histplot(means,kde=True)
```
