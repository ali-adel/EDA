# Two tailed

```python
D = np.concatenate((np.ones(530),np.zeros(470))) # our sample
print(D.mean()) # sample mean
ztest(D,value=0.5,alternative="two-sided") # returns  z-critical and p-value
# so p-value > alpha we will not reject null hypothesis(H0)
```

# Right tailed

```python
D = np.concatenate((np.ones(530),np.zeros(470))) # our sample
print(D.mean()) # sample mean
ztest(D,value=0.5,alternative="larger") # returns  z-critical and p-value
# so p-value < alpha we will reject null hypothesis(H0)
```

# left tailed

```python
D = np.concatenate((np.ones(530),np.zeros(470))) # our sample
print(D.mean()) # sample mean
ztest(D,value=0.5,alternative="smaller") # returns  z-critical and p-value
# so p-value > alpha we will not reject null hypothesis(H0)
```

# when decrease the ones size the sample mean decrease then p-value decrease

```python
D = np.concatenate((np.ones(470),np.zeros(530))) # our sample
print(D.mean()) # sample mean
ztest(D,value=0.5,alternative="smaller") # returns  z-critical and p-value
# so p-value < alpha we will  reject null hypothesis(H0)
```

# **Using t-test**

# **Ex:** houses and it's areas
**H0 : mean = 5000**

# Two sided

```python
D = { "areas" : [1000,1500,2300,3540,4120,4560,5490,3460,4750,2300,9000,8600,7100]}
df = pd.DataFrame(D)
print(df["areas"].mean())
ttest_1samp(a=df,popmean=5000)# p-value for  two sided p-val = 0.44346471
```

# Left sided

```python
ttest_1samp(a=df,popmean=5000,alternative="less")# p-value for  left sided p-val = 0.22173235 or (p-val for two-sided / 2)
```

# Right sided

```python
ttest_1samp(a=df,popmean=5000,alternative="greater",)# p-value for  right sided p-val = 0.77826765 as the sample mean =  4440.0 (left hypothesis mean)
```
