# Step Three

# We plot the histogram depending on the freqency table

```python
  # we will make the bin size here  5

hist = sns.histplot(data,bins=5)
```

# what if we cahnge the bin size

```python
# bin size here is 9
hist = sns.histplot(data,bins=9)
```

# what if we want to make the bin just for a unique number

```python
hist = sns.histplot(data,discrete=True)
# we notice that 15  the most frequent number
```

# to get the estimated shape of data

```python
hist = sns.histplot(data,kde=True)
```

# to get the probability

```python
hist = sns.histplot(data,kde=True,stat='probability')
```

# to get the denisty or pdf

```python
hist = sns.histplot(data,kde=True,stat='density')
```
