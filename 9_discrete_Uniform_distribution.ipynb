```python
import numpy as np
from scipy.stats import randint
import seaborn as sns
```

# Uniform_distribution

# each RV  is equaly likely for example rolling dice experiment  each value of RV from 1 to 6 have probability = 1/6

image.png

**a => start of RV**

**b => end of RV**

```python
a = 1
b = 6
x = 2
prob_2 = randint.pmf(x,a,b+1)
prob_2 # probability of 2
```

# Theoritical

```python
a = 1
b = 6
x = 5
prob_5 = randint.pmf(x,a,b+1)
prob_5 # probability of 5
```

```python
a = 1
b = 6
x = np.arange(1,b+1)
prob = randint.pmf(x,a,b+1)
prob # probability of each RV x = 0.16666667
```

```python
print("mean is : " , x.mean(),"or (a+b)/2 : ", (a+b)/2)
```

```python
dist = sns.barplot(x=x,y=prob)
```

# Practical (sample)

```python
prob_s = np.random.randint(a,b+1,size=10)
print("mean : ",prob_s.mean(),"\n")
h_plot= sns.histplot(prob_s,discrete=True)
```

```python
prob_s = np.random.randint(a,b+1,size=100)
print("mean : ",prob_s.mean(),"\n")
h_plot = sns.histplot(prob_s,discrete=True)
```

```python
prob_s = np.random.randint(a,b+1,size=1000)
print("mean : ",prob_s.mean(),"\n")
h_plot = sns.histplot(prob_s,discrete=True)
```
